```python
#  Sobel operator applied in a chosen direction (x or y)
def abs_sobel_thresh(img, orient='x', thresh_min=0, thresh_max=255):

    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    if orient == 'x':
        sobel_dir = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    else:
        sobel_dir = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    sobel_abs = np.absolute(sobel_dir)

    max_sobel = np.max(sobel_abs)
    scaled = np.uint8(255*(sobel_abs/max_sobel))

    mask = np.zeros_like(scaled)
    mask[(scaled > thresh_min) &
         (scaled < thresh_max)]  = 1

    return mask

# Take the magnitude of the gradient
def mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)):

    # Apply the following steps to img
    # 1) Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # 2) Take the gradient in x and y separately
    sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # 3) Calculate the magnitude
    magnitude = np.sqrt((sobel_x ** 2) + (sobel_y ** 2))

    # 4) Scale to 8-bit (0 - 255) and convert to type = np.uint8
    scaled  = np.uint8(magnitude*255/np.max(magnitude))
    # 5) Create a binary mask where mag thresholds are met
    mask = np.zeros_like(scaled)
    mask[(scaled > mag_thresh[0]) & (scaled < mag_thresh[1])] = 1
    # 6) Return this mask as your binary_output image

    binary_output = mask

    return binary_output

# Take the direction of the gradient
def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi / 2)):
    # Apply the following steps to img
    # 1) Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # 2) Take the gradient in x and y separately
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # 3) Take the absolute value of the x and y gradients
    absx = np.absolute(sobelx)
    absy = np.absolute(sobely)
    # 4) Use np.arctan2(abs_sobely, abs_sobelx) to calculate the direction of the gradient
    direction = np.arctan2(absy, absx)
    # 5) Create a binary mask where direction thresholds are met
    mask = np.zeros_like(direction)
    mask[(direction > thresh[0]) & (direction < thresh[1])] = 1
    # 6) Return this mask as your binary_output image

    return mask

def hls_select(img, channel='s', thresh=(0, 255)):
    # 1) Convert to HLS color space
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    # 2) Apply a threshold to the S channel
    if channel == 'h':
        ch = 0
    elif channel == 'l':
        ch = 1
    elif channel == 's':
        ch = 2
    
    chan = hls[:, :, ch]
#     s = hls[:, :, 2]

    # 3) Return a binary image of threshold result
    mask = np.zeros_like(chan)
    mask[(chan > thresh[0]) & (chan <= thresh[1])] = 1

    binary_output = mask
    return binary_output

```
